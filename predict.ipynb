```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read Data
jungnangcheon_df = pd.read_excel('./readonly/Data.xlsx')
date = jungnangcheon_df['Date']
y    = jungnangcheon_df['BOD(mg/L)']
X    = jungnangcheon_df[['Temperature (°C)', 'Precipitation (mm)', 'Air Pressure (hPa)', 'Solar Radiation (MJ/m2)']]
# print(jungnangcheon_df.keys())
```

```python
# Plot input data
import matplotlib.dates as dates
fig  = plt.figure()
fig  .suptitle('jungnangcheon River', fontsize=16)
temp = fig.add_subplot()
temp .set_ylabel('T (ºC)')
temp .set_ylim(0, 30)
plt.gca().xaxis.set_major_locator(dates.DayLocator(interval=3))
plt.gcf().autofmt_xdate()
temp .plot(date, X['Temperature (°C)'], '--bo', label = 'Air Temperature', markersize=4)
```
